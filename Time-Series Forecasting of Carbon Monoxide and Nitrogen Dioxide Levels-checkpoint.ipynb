### => Outliers Detection <==

#### Outlier Detection, where we will identify and handle extreme values in CO(GT) and NO₂(GT) using statistical thresholds 

```python
import numpy as np
```

```python
# Define a function to remove outliers using the IQR method
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # First quartile (25%)
    Q3 = data[column].quantile(0.75)  # Third quartile (75%)
    IQR = Q3 - Q1                     # Interquartile range
    lower_bound = Q1 - 1.5 * IQR      # Lower bound
    upper_bound = Q3 + 1.5 * IQR      # Upper bound
    # Remove or cap outliers
    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])
    return data
```

```python
# Apply outlier removal on 'CO(GT)' and 'NO₂(GT)'
df = remove_outliers_iqr(df, 'CO(GT)')
df = remove_outliers_iqr(df, 'NOx(GT)')
```

###  Step : 1 Check stationarity of NO₂(GT)

#### Let's first visualize the time series and perform the ADF test to check stationarity.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

# Check column names
print(df.columns)

# Plot CO(GT) time series
plt.figure(figsize=(10, 6))
plt.plot(df['CO(GT)'], label="CO(GT)")
plt.title("CO(GT) Time Series")
plt.xlabel("Time")
plt.ylabel("CO(GT)")
plt.legend()
plt.show()

# Perform ADF Test on CO(GT)
adf_test_co = adfuller(df['CO(GT)'].dropna())  # Dropping NaN values for test
print("ADF Test Statistic (CO):", adf_test_co[0])
print("p-value (CO):", adf_test_co[1])
if adf_test_co[1] <= 0.05:
    print("The CO(GT) series is stationary.")
else:
    print("The CO(GT) series is not stationary.")

# Check if 'NO₂(GT)' exists, and then plot and perform ADF Test
if 'NO₂(GT)' in df.columns:
    # Plot NO₂(GT) time series
    plt.figure(figsize=(10, 6))
    plt.plot(df['NO₂(GT)'], label="NO₂(GT)", color="orange")
    plt.title("NO₂(GT) Time Series")
    plt.xlabel("Time")
    plt.ylabel("NO₂(GT)")
    plt.legend()
    plt.show()

    # Perform ADF Test on NO₂(GT)
    adf_test_no2 = adfuller(df['NO₂(GT)'].dropna())  # Dropping NaN values for test
    print("ADF Test Statistic (NO₂):", adf_test_no2[0])
    print("p-value (NO₂):", adf_test_no2[1])
    if adf_test_no2[1] <= 0.05:
        print("The NO₂(GT) series is stationary.")
    else:
        print("The NO₂(GT) series is not stationary.")
else:
    print("The column 'NO₂(GT)' is not found in the dataset.")
```

#### If the series is not stationary:
##### We'll apply differencing to make it stationary and re-run the ADF test.

```python
# Apply differencing (if needed)
df['CO(GT)_diff'] = df['CO(GT)'].diff()

# Plot differenced series
plt.figure(figsize=(10, 6))
plt.plot(df['CO(GT)_diff'], label="Differenced CO(GT)")
plt.title("Differenced CO(GT) Time Series")
plt.xlabel("Time")
plt.ylabel("Differenced CO(GT)")
plt.legend()
plt.show()

# Re-run ADF Test
adf_test_diff = adfuller(df['CO(GT)_diff'].dropna())
print("ADF Test Statistic after differencing:", adf_test_diff[0])
print("p-value:", adf_test_diff[1])
if adf_test_diff[1] <= 0.05:
    print("The series is now stationary.")
else:
    print("The series is still not stationary.")
```

#### Step 2 : Identify ARIMA (p, d, q) using ACF/PACF plots
##### Use the ACF (q) and PACF (p) plots.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF and PACF for CO(GT)
plt.figure(figsize=(12, 6))

# ACF Plot
plt.subplot(121)
plot_acf(df['CO(GT)_diff'].dropna(), ax=plt.gca(), lags=30)
plt.title("ACF Plot for CO(GT)")

# PACF Plot
plt.subplot(122)
plot_pacf(df['CO(GT)_diff'].dropna(), ax=plt.gca(), lags=30)
plt.title("PACF Plot for CO(GT)")

plt.tight_layout()
plt.show()
```

### Step 3 :  Fit ARIMA Model
#### Using the chosen parameters (p, d, q) from the above analysis, fit an ARIMA model.

```python
from statsmodels.tsa.arima.model import ARIMA

# Set ARIMA order based on ACF/PACF (example: p=1, d=1, q=1)
arima_order = (1, 1, 1)

# Assuming your DataFrame is named 'df', fit the ARIMA model
arima_model = ARIMA(df['CO(GT)'].dropna(), order=arima_order)  # Use 'df' instead of 'data'
arima_result = arima_model.fit()

# Summary of the model
print(arima_result.summary())
```
